# Perform some analysis on the datset

```python
import numpy as np
import matplotlib.pyplot as plt  
import random
```

```python
# load dataset
dataset_file = "dataset_career_con_2019.npz"
dataset = np.load(dataset_file, allow_pickle=True)
x_train = dataset["x_train"]
x_test = dataset["x_test"]
y_train = dataset["y_train"]
y_test = dataset["y_test"]
x_feature_names = dataset["x_feature_names"]
y_category_names = dataset["y_category_names"]
```

```python
# show size of train vs. test part
train_size = len(x_train)
test_size = len(x_test)
print("Number of train samples: {}, number of test samples: {}".format(train_size, test_size))
```

```python
# show shape of x_train, and feature names
print("Number of sequences: {}, sequence length: {}, number of features: {}".format(*x_train.shape))
print("Features of X: {}".format(x_feature_names))
```

```python
# show some details about the labels
print("Label categories: {}".format(y_category_names))
label_frequencies_train = np.apply_along_axis(sum, 0, y_train)
label_frequencies_test = np.apply_along_axis(sum, 0, y_test)
x_axis = np.arange(len(y_category_names))
plt.bar(x_axis - 0.2, label_frequencies_train, 0.4, label = 'Train')
plt.bar(x_axis + 0.2, label_frequencies_test, 0.4, label = 'Test')
plt.xticks(x_axis, y_category_names, rotation='vertical')
plt.xlabel("Label")
plt.ylabel("Frequency")
plt.title("Label frequency")
plt.tight_layout()
plt.legend()
plt.show()
```

```python
# show time series of all features for 1 sample
plt.figure()
one_time_series = x_train[random.randint(0, x_train.shape[1])]
time = range(len(one_time_series))
for i in range(one_time_series.shape[1]):
    plt.plot(time, one_time_series[:,i])
plt.xlabel("Time")
plt.ylabel("Feature values")
plt.title("One time series, different features")
plt.show()
```
